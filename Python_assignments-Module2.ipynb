```python
"""
    Read the pickle file 'Data/Stress.pkl' and transform the contents into a
    a pandas DataFrame of the following form:
    
            +------+------+------+-----+------+-------+
            | ACCx | ACCy | ACCz | ECG | RESP | LABEL |
            +------+------+------+-----+------+-------+
            |      |      |      |     |      |       |
            +------+------+------+-----+------+-------+
            |      |      |      |     |      |       |
            +------+------+------+-----+------+-------+
            |      |      |      |     |      |       |
            +------+------+------+-----+------+-------+

            Shape: (700, 6)

    Save this DataFrame as a CSV file.

    Modules
    -------
    pickle, pandas
    
    Lookups
    -------
    NumPy and pandas basics (Check the Lookups folder)
    Try to complete the tasks in the Lookups, and play around with the notebooks.
"""        
```

```python
def pick_to_panda():
    new_file = open('/home/legion/Desktop/python_module2/Stress.pkl', 'rb')
    db = {}
    dp = {}
    d = {}
    db = pkl.load(new_file)
    dp = pd.DataFrame()
    d = db['signal']
    acc = d['ACC']
    ecg = [i[0] for i in d['ECG']]
    resp = [i[0] for i in d['RESP']]
    label = [i[0] for i in d['LABEL']]
    ACCx = [i[0] for i in acc]
    ACCy = [i[1] for i in acc]
    ACCz = [i[2] for i in acc]
    dp['ACCx'] = ACCx
    dp['ACCy'] = ACCy
    dp['ACCz'] = ACCz
    dp['ECG'] = ecg
    dp['RESP'] = resp
    dp['LABEL'] = label
    dp.to_csv("PandaData.csv", index=False, encoding='utf-8')
    
```

```python
"""
    Find the count of each value in `LABEL`.
    
    Add column `BINARY` to the DataFrame which contains 1 if `LABEL` is 2,
    else 0.
    
    Find the mean, minimum, maximum and standard deviation of `RESP` for each value in `LABEL`.
"""
```

```python
# Your function goes here
def Math_func():
    new_file = open('/home/legion/Desktop/python_module2/Stress.pkl', 'rb')
    db = {}
    dp = {}
    d = {}
    db = pkl.load(new_file)
    dp = pd.DataFrame()
    d = db['signal']
    acc = d['ACC']
    ecg = [i[0] for i in d['ECG']]
    resp = [i[0] for i in d['RESP']]
    label = [i[0] for i in d['LABEL']]
    ACCx = [i[0] for i in acc]
    ACCy = [i[1] for i in acc]
    ACCz = [i[2] for i in acc]
    dp['ACCx'] = ACCx
    dp['ACCy'] = ACCy
    dp['ACCz'] = ACCz
    dp['ECG'] = ecg
    dp['RESP'] = resp
    dp['LABEL'] = label
    binary = list(map(lambda x: int(x == 2), label))
    dp['BINARY'] = binary
    maximum = dp['RESP'].groupby(dp['LABEL']).max()
    minimum = dp['RESP'].groupby(dp['LABEL']).min()
    mean = dp['RESP'].groupby(dp['LABEL']).mean()
    stdev = dp['RESP'].groupby(dp['LABEL']).std()
    print(maximum)
    print(minimum)
    print(mean)
    print(stdev)
```

```python
# Your function goes here
def plots():
    new_file = open('/home/legion/Desktop/python_module2/Stress.pkl', 'rb')
    db = {}
    dp = {}
    d = {}
    db = pkl.load(new_file)
    dp = pd.DataFrame()
    d = db['signal']
    acc = d['ACC']
    colour = ['r', 'g', 'b', 'y', 'c', 'm', 'k']
    ecg = [i[0] for i in d['ECG']]
    resp = [i[0] for i in d['RESP']]
    label = [i[0] for i in d['LABEL']]
    ACCx = [i[0] for i in acc]
    ACCy = [i[1] for i in acc]
    ACCz = [i[2] for i in acc]
    dp['ACCx'] = ACCx
    dp['ACCy'] = ACCy
    dp['ACCz'] = ACCz
    dp['ECG'] = ecg
    dp['RESP'] = resp
    dp['LABEL'] = label
    binary = list(map(lambda x: int(x == 2), label))
    dp['BINARY'] = binary
    col = [colour[i] for i in binary]
    for i in range(len(resp)):
        plt.scatter(binary[i], resp[i], c=col[i])
    plt.show

```

Some simple file operations.

```python
"""
    Generate a text file containing 50 integer elements between 1 and 20 (use `random`). 
    Identify the unique numbers and the number of their occurrences, and write this into 
    a new text file.
    
    Example
    -------
    
    Generated File
    --------------
    1 2 3 4 3 2 1 ...
    
    Count File
    ----------
    1 - 2
    2 - 2
    3 - 2
    4 - 1
    ...
"""
```

```python
def F():
    random_list = list()
    histo = dict()
    f = open("Occurences.txt", "w+")
    for i in range(0, 50):
        random_list.append(random.randrange(0, 21, 1))
    for i in random_list:
        if i not in histo:
            histo[i] = 1
        else:
            histo[i] += 1
    f.write(str(histo))
    f.close()
    f = open("Occurences.txt", "r")
    print(f.read())
    
```

```python
# Call your function here
F()
```
